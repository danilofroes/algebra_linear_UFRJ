```python
import numpy as np
import matplotlib.pyplot as plt

u = np.array([1, 2, 0])
v = np.array([0, 1, 1])

pontos_x, pontos_y, pontos_z = [], [], []
alpha, beta = None, None

qntd_pontos = 500
np.random.seed(42)

for _ in range(qntd_pontos):
    alpha = np.random.uniform(-10, 10)
    beta = np.random.uniform(-10, 10)

    p = (alpha * u) + (beta * v)

    pontos_x.append(p[0])
    pontos_y.append(p[1])
    pontos_z.append(p[2])
```

```python
fig = plt.figure(figsize = (10, 8))
ax = fig.add_subplot(111, projection = '3d')

ax.scatter(pontos_x, pontos_y, pontos_z, c = 'gray', alpha = 0.3, s = 5, label = 'Span (0 Plano)')

ax.scatter([0], [0], [0], color = 'black', s = 100, marker = 'o', label = 'Origem')

ax.quiver(0, 0, 0, u[0], u[1], u[2], color = 'red', length = 10, label = 'Vetor u', arrow_length_ratio = 0.1)
ax.quiver(0, 0, 0, v[0], v[1], v[2], color = 'blue', length = 10, label = 'Vetor v', arrow_length_ratio = 0.1)

ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')
ax.set_title(r'Visualizao do Span: $Span\{\vec{u}, \vec{v}\}$')
ax.legend()

plt.show()
```
